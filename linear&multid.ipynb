```python
%matplotlib inline
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

lr = LinearRegression(fit_intercept = True)

x = boston.data[:, 5]
y = boston.target
x = np.transpose(np.atleast_2d(x))
lr.fit(x, y)

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.set_xlabel("# of rooms per dwellings")
ax.set_ylabel("House Price")
xmin = x.min()
xmax = x.max()
ax.plot([xmin, xmax], 
        [lr.predict(np.transpose(np.atleast_2d(xmin))), lr.predict(np.transpose(np.atleast_2d(xmax)))], 
        '-', lw = 2)
ax.scatter(x, y, s = 2)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, lr.predict(x))
print("MSE of Training data is: {:.3}".format(mse))
rmse = np.sqrt(mse)
print("RMSE of Training data is: {:.3}".format(rmse))
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, lr.predict(x))
print("R2 Score on the training data is: {:.3}".format(r2))

print("using score method: {:.3}".format(lr.score(x,y)))
```

```python
#multi-dimensional regression
x = boston.data
y = boston.target

lr.fit(x, y)
#can't plot coz using all the attributes and we have 14-d regression hyperplane

mse = mean_squared_error(y, lr.predict(x))
print("When using all the attributes we get MSE of: {:.3}".format(mse), "and RMSE of: {:.3}".format(np.sqrt(mse)))
print("The R2_Score is: {:.3}".format(r2_score(y, lr.predict(x))))
```

```python
#plotting the predicted values to the actual values

predicted_values = lr.predict(x)
fig, ax = plt.subplots()

ax.scatter(predicted_values, y, s=2)
ax.plot([y.min(), y.max()], [y.min(), y.max()], ':',  lw = 2)
ax.set_ylabel("True values")
ax.set_xlabel("Predicted Values")
```

```python
from sklearn import model_selection

#kf = KFold(n_splits=5)
p = model_selection.cross_val_predict(lr, x, y, cv= model_selection.KFold(n_splits=5))
rmse_cv = np.sqrt(mean_squared_error(p, y))
print('RMSE on 5-fold CV: {:.2}'.format(rmse_cv))
print('R2_Score: {:.3}'.format(r2_score(y, p)))
```
